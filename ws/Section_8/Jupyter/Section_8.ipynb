## Lecture 68 - Example: Geocoding Addresses with Pandas and Geopy

```python
import geopy
from geopy.geocoders import Nominatim
nom=Nominatim()
```

```python
nom.geocode("3995 23rd St, San Francisco, CA 94114")
```

```python
nom.geocode("Czstochowska 8, 02-344 Warszawa, Polska")
```

```python
n=nom.geocode("Salon Play")
n
```

```python
print(n.latitude)
print(n.longitude)
```

Pandas has methods to automatically use given method of function on all records of DataFrame, without need for writing for loop

```python
df1["Coordinates"]=df1["Address"].apply(nom.geocode)
df1
```

```python
df1["x"]=df1["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df1["y"]=df1["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df1
```
